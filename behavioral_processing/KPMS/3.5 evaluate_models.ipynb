```python
import os
import numpy as np
import keypoint_moseq as kpms
```

```python
PROJECT_DIR = "/mnt/d/Project/Keypoint-Moseq/"
```

```python
model_name_0 = 'my_models-0'
model_name_1 = 'my_models-1'

results_1 = kpms.load_results(PROJECT_DIR, model_name_0)
results_2 = kpms.load_results(PROJECT_DIR, model_name_1)

kpms.plot_confusion_matrix(results_1, results_2)
```

```python
model_names = [f for f in os.listdir(PROJECT_DIR) if "my_models-" in f]

eml_scores, eml_std_errs = kpms.expected_marginal_likelihoods(PROJECT_DIR, model_names)
best_model = model_names[np.argmax(eml_scores)]
print(f"Best model: {best_model}")

kpms.plot_eml_scores(eml_scores, eml_std_errs, model_names)
```
