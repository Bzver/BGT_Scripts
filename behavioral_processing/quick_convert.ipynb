```python
import os
import glob
import shutil

folder = r"D:\Data\Videos\ASOiD Predict\processed"
tabletable = [
    ("20250913", 3, "D", "SE"),
    ("20250918", 3, "L", "SE"),
    ("20251012", 4, "D", "SE"),
    ("20251012", 3, "L", "VG"),
]

os.makedirs(os.path.join(folder, "processed"), exist_ok=True)

for date, floor, _, sex in tabletable:
    yyyy = date[:4]
    mmdd = date[4:]
    files = glob.glob(os.path.join(folder, f"{mmdd}_*_{floor}0*"))
    for f in files:
        fname = os.path.splitext(os.path.basename(f))[0]
        dom = fname[5:8]
        day = fname[9:13]
        fid = fname[14:17]
        other = fname[17:]
        new_name = f"{yyyy}{mmdd}_{dom}_{day}_{fid}_{sex}{other}"
        new_path = os.path.join(folder, "processed", new_name)
        shutil.move(f, new_path)
```
