```python
import os
import glob
import shutil

folder = r"D:\Project\ASOID-Models\May-01-2026\videos"
tabletable = [
    ("20250913", 3, "D", "SE"),
    ("20250918", 3, "L", "SE"),
    ("20251012", 4, "D", "SE"),
    ("20251012", 3, "L", "VG"),
    ("20251012", 2, "D", "SE"),
    ("20251018", 3, "D", "SE"),
    ("20251018", 2, "L", "SE"),
    ("20251101", 4, "D", "VG"),
    ("20251101", 3, "L", "VG"),
    ("20251101", 2, "L", "VG"),
    ("20251101", 1, "L", "VG"),
    ("20251117", 4, "D", "SE"),
    ("20251117", 3, "D", "SE"),
    ("20251117", 2, "L", "SE"),
    ("20251201", 3, "L", "SE"),
    ("20251201", 1, "L", "VG"),
    ("20260324", 4, "D", "SE"),
    ("20260324", 2, "D", "SE"),
    ("20260416", 3, "D", "SE"),
    ("20260416", 2, "L", "SE"),
]

os.makedirs(os.path.join(folder, "processed"), exist_ok=True)

for date, floor, _, sex in tabletable:
    yyyy = date[:4]
    mmdd = date[4:]
    files = glob.glob(os.path.join(folder, f"{mmdd}_*_{floor}0*"))
    print(files)
    for f in files:
        fname = os.path.basename(f)
        dom = fname[5:8]
        day = fname[9:13]
        fid = fname[14:17]
        other = fname[17:]
        new_name = f"{yyyy}{mmdd}_{dom}_{day}_{fid}_{sex}{other}"
        new_path = os.path.join(folder, "processed", new_name)
        if os.path.isfile(new_path):
            print(f"Warning: {new_path} already exists!")
        else:
            shutil.move(f, new_path)
```
